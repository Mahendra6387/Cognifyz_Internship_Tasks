```python
import pandas as pd
df = pd.read_csv("task.csv") 
print(df.head())
print(df.columns)
top_cuisines = df['Cuisines'].value_counts().head(3)
print("Top 3 Most Common Cuisines", top_cuisines)
total_restaurants = len(df)
percentages = (top_cuisines / total_restaurants) * 100
percentages = percentages.round(2)  
print("Percentage of restaurants serving each top cuisine", percentages)
```

```python
import pandas as pd
data = pd.read_csv("task.csv")
city_counts = data['City'].value_counts()
top_city = city_counts.idxmax()
top_city_count = city_counts.max()
print(f"City with the highest number of restaurants: {top_city} ({top_city_count} restaurants)\n")
avg_rating_city = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
print("Average Rating per City", avg_rating_city.head(), "\n")
best_rated_city = avg_rating_city.idxmax()
highest_avg_rating = avg_rating_city.max()
print(f"City with the highest average rating: {best_rated_city} (highest_avg_rating:.2f)")
```

```python
df = pd.read_csv("task.csv")
delivery_counts = df['Has Online delivery'].value_counts()
total = delivery_counts.sum()
percent_online = (delivery_counts['Yes'] / total) * 100
percent_offline = (delivery_counts['No'] / total) * 100
print(f"Percentage offering online delivery: {percent_online:.2f}%")
print(f"Percentage not offering online delivery: {percent_offline:.2f}%")
avg_rating_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
avg_rating_no_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()
print(f"\nAverage rating (Online Delivery): {avg_rating_delivery:.2f}")
print(f"Average rating (No Online Delivery): {avg_rating_no_delivery:.2f}")
```
